```python
class Electrical_system_information:
    
    def __init__(self, volAve, volVar, currentAve, currentVar, pfAve, pfVar):
        
        self.volAve = volAve
        self.volVar = volVar
        self.currentAve = currentAve
        self.currentVar = currentVar
        self.pfAve = pfAve
        self.pfVar = pfVar
        
    def voltage(self):
        
        return np.random.normal(loc=self.volAve, scale =self.volVar , size = 1)
        
    def current(self):
        
        return np.random.normal(loc=self.currentAve, scale =self.currentVar , size = 1)
    
    def powerfarctor(self):
        
        return np.random.normal(loc=self.pfAve, scale =self.pfVar , size = 1)
    
```

```python
class Environmental_sensors:
    def __init__(self, TempAve, TempVar, humidityAve, humidityVar):
        
        self.TempAve = TempAve
        self.TempVar = TempVar
        self.humidityAve = humidityAve
        self.humidityVar = humidityVar
               
    def fan_status(self):
        
        return random.choice([True, False])
    
    def room_temp(self):
        
        return np.random.normal(loc=self.TempAve, scale =self.TempVar , size = 1)
    
    def room_humidity(self):
        
        return np.random.normal(loc=self.humidityAve, scale =self.humidityVar , size = 1)
    
```
